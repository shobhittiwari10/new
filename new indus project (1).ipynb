```
#  FIRST LOAD THE DATA FROM SOURCES
    
    
 from pyspark.sql import SparkSession
spark=SparkSession.builder.appName('appmane').getOrCreate()
df1=spark.read.csv('/FileStore/tables/AppleStore1-3.csv',header=True,inferSchema=True)
df1.show()
```

```
df2=spark.read.csv('/FileStore/tables/AppleStore2-5.csv',header=True,inferSchema=True)
df2.show()
```

```
df3=spark.read.csv('/FileStore/tables/AppleStore3-1.csv',header=True,inferSchema=True)
df3.show()
```

```
df4=spark.read.csv('/FileStore/tables/AppleStore4-1.csv',header=True,inferSchema=True)
df4.show()
```

```
columns = df1.columns

df2 = df2.select(columns)
df3 = df3.select(columns)
df4 = df4.select(columns)

merged_df = df1.union(df2).union(df3).union(df4)
merged_df.show()
```
